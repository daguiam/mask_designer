```python

#initialize
layout = pya.Layout()
top = layout.create_cell('top')



# single_calibration = layout.create_cell('rectangle')
# layer_n = layout.layer("layer_n")


layers = np.arange(0,128)




# Create matrix of squares for measurement
cell_gs_matrix = layout.create_cell('gs_matrix')

step = 50.0
square_side = 30

rows = range(8)
cols = range(16)

CC,RR = np.meshgrid(cols, rows, indexing='xy')

for i, (layer, col, row) in enumerate(zip(layers, CC.flatten(), RR.flatten())):
    # print(i, layer, row, col)
    layer_name = "layer%03d"%(layer)

    layer_i = layout.layer(layer_name)

    x,y = step*col, -step*row

    # print(x,y, x+square_side, y+square_side)


    # new_instance = pya.CellInstArray(cell_single_level.cell_index(), trans, pya.Vector(50, 0 ), pya.Vector(0, 50), 16, 8 )
    # cell_single_level.shapes(layer_i).insert(pya.Box(square_side, square_side).move(pya.Vector(dx, dy)))
    # cell_single_level.shapes(layer_i).insert(pya.DBox(square_side, square_side))
    cell_gs_matrix.shapes(layer_i).insert(pya.DBox(x,y, x+square_side, y+square_side))

trans = pya.Trans(pya.Point(0,0))

new_instance = pya.DCellInstArray(cell_gs_matrix.cell_index(), trans, pya.Vector(50, 0 ), pya.Vector(0, 50), 1,1)
# reference cell index, transformation, x offset, y offset, number of x instances, and number of y instances

# calibration_instance = pya.Instance()
top.insert(new_instance)

total_width = 780



## Create the and decreasing increasing GS ramps
cell_ramps = layout.create_cell('gs_ramps')


ramp_width = 5
ramp_height = 50
step_gs = ramp_width

x_position = 70.0
y_position = 50.0

# decreasing_ramp:
for i, (layer) in enumerate(layers):

    layer_name = "layer%03d"%(layer)

    layer_i = layout.layer(layer_name)
    x,y = x_position+step_gs*i, y_position+ 0

    cell_ramps.shapes(layer_i).insert(pya.DBox(x,y, x+ramp_width, y+ramp_height))

# add start and ending squares
layer_i = layout.layer("layer%03d"%(127))
cell_ramps.shapes(layer_i).insert(pya.DBox(0, y_position, 60, y_position+ramp_height))
cell_ramps.shapes(layer_i).insert(pya.DBox(total_width-60.0,y_position, total_width, y_position+ramp_height))

trans = pya.Trans(pya.Point(x_position,y_position))
trans = pya.Trans(pya.Point(0,0))

new_instance = pya.DCellInstArray(cell_ramps.cell_index(), trans, pya.Vector(50, 0 ), pya.Vector(0, 50), 1,1)

top.insert(new_instance)



## Create the increasing and decreasing increasing GS ramps
cell_triangle = layout.create_cell('gs_triangle]')



ramp_width = 2
ramp_height = 50
step_gs = ramp_width

x_position = (780-ramp_width*len(layers)*2)/2
y_position = 120.0

y_position2 = 190
x_position2 = total_width/2

end_x_position = x_position+ramp_width*len(layers)*2

# decreasing_ramp:
for i, (layer) in enumerate(layers):

    layer_name = "layer%03d"%(layer)

    layer_i = layout.layer(layer_name)
    x,y = x_position+step_gs*i, y_position+ 0

    cell_triangle.shapes(layer_i).insert(pya.DBox(x,y, x+ramp_width, y+ramp_height))
    
    x,y = end_x_position-(step_gs*(i+1)), y_position+ 0
    print(x)
    cell_triangle.shapes(layer_i).insert(pya.DBox(x,y, x+ramp_width, y+ramp_height))

    # increasing
    x,y = x_position2+step_gs*i, y_position2+ 0

    cell_triangle.shapes(layer_i).insert(pya.DBox(x,y, x+ramp_width, y+ramp_height))
    
    x,y = x_position2-(step_gs*(i+1)), y_position2+ 0
    
    cell_triangle.shapes(layer_i).insert(pya.DBox(x,y, x+ramp_width, y+ramp_height))



# add start and ending squares
layer_i = layout.layer("layer%03d"%(127))
cell_triangle.shapes(layer_i).insert(pya.DBox(0, y_position, 100, y_position+ramp_height))
cell_triangle.shapes(layer_i).insert(pya.DBox(total_width-100.0,y_position, total_width, y_position+ramp_height))


cell_triangle.shapes(layer_i).insert(pya.DBox(0, y_position2, 100, y_position2+ramp_height))
cell_triangle.shapes(layer_i).insert(pya.DBox(total_width-100.0,y_position2, total_width, y_position2+ramp_height))

trans = pya.Trans(pya.Point(x_position,y_position))
trans = pya.Trans(pya.Point(0,0))

new_instance = pya.DCellInstArray(cell_triangle.cell_index(), trans, pya.Vector(50, 0 ), pya.Vector(0, 50), 1,1)


top.insert(new_instance)



layout.write("test.dxf")



layout.write("gs_calibration_v20240125.dxf")
```

```python

#initialize
layout = pya.Layout()
top = layout.create_cell('top')
c_circle = layout.create_cell('circle')
l_circle = layout.layer("layer000")

# make a circle pcell
radius = 200 #radius in nm
nr_points = 32 #number of points
#create an array of angles
angles = np.linspace(0,2*np.pi,nr_points+1)[0:-1]
points = [] #array of point
for angle in angles:
    points.append(pya.Point(radius*np.cos(angle),radius*np.sin(angle)))
circle = pya.SimplePolygon(points)



#for hexagonal grid
a = 400 #hexagon radius towards flat
#number of instances
i=10
c_circle.shapes(l_circle).insert(circle)
c_circle.shapes(l_circle).insert(circle.moved(a*np.sqrt(3),a))



trans = pya.Trans(pya.Point(0,0))
new_instance = pya.CellInstArray(c_circle.cell_index(),trans,pya.Vector(2*a*np.sqrt(3), 0 ), pya.Vector(0, 2*a), i/2, i/2)
top.insert(new_instance)
layout.write("test.dxf")
```
